**NUMBER OF GOALS SCORED BY TEAMS THAT PLAYED IN THE WORLDCUP**

```python
total_goals_team1 = a.groupby('team1')['number of goals team1'].sum()
total_goals_team2 = a.groupby('team2')['number of goals team2'].sum()
total_goals = pd.concat([total_goals_team1, total_goals_team2], axis=1)
total_goals
```

**TEAMS AND THEIR POSSESSION IN THEIR GAMES THROUGHTOUT THE WORLDCUP**

```python
a['possession team1'] = a['possession team1'].astype(str).str.replace('%', '').astype(float) # Convert to string type before using .str
a['possession team2'] = a['possession team2'].astype(str).str.replace('%', '').astype(float)

avg_possession_team1 = a.groupby('team1')['possession team1'].mean()
avg_possession_team2 = a.groupby('team2')['possession team2'].mean()
avg_possession = pd.concat([avg_possession_team1, avg_possession_team2], axis=1)
avg_possession
```

**THIS SHOWS THE GOAL DIFFERENCE THE TEAMS WHICH ALSO AFFECTS THE QUALIFICATIONS DURING THE WORLDCUP**

```python
a['goal_difference'] = a['number of goals team1'] - a['number of goals team2']
a['goal_difference']
```

**THIS GRAPH SHOWS WHICH TEAM SCORED THE MOST GOALS DURING THE TOURNAMENT**

```python
total_goals_team1 = a.groupby('team1')['number of goals team1'].sum()
total_goals_team2 = a.groupby('team2')['number of goals team2'].sum()
total_goals = pd.concat([total_goals_team1, total_goals_team2], axis=1).fillna(0)
total_goals['total_goals'] = total_goals.sum(axis=1)
total_goals = total_goals.sort_values('total_goals', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=total_goals['total_goals'], y=total_goals.index)
plt.title('Total Goals Scored by Each Team')
plt.xlabel('Total Goals')
plt.ylabel('Teams')
plt.show()
```

**THIS BOXPLOT SHOWS US THAT THE POSSESION PERCENTAGE BETWEEN TEAM1 AND TEAM 2**

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=a[['possession team1', 'possession team2']].apply(pd.to_numeric), orient='h')
plt.title('Distribution of Possession Percentage for Team1 vs Team2')
plt.xlabel('Possession Percentage')
plt.show()
```

**THIS HISPLOT SHOWS THE DISTRIBUTION OF GOAL DIFFERENCES WHICH HELPS US UNDERSTAND WHICH TEAM PALYS MORE AGGRESIVE**

```python
a['goal_difference'] = a['number of goals team1'] - a['number of goals team2']

plt.figure(figsize=(8,6))
sns.histplot(a['goal_difference'], kde=True, bins=20)
plt.title('Distribution of Goal Differences')
plt.xlabel('Goal Difference (Team1 - Team2)')
plt.ylabel('Frequency')
plt.show()
```

**THIS SCATTERPLOT SHOWS US THAT THE POSSESSION PERCENTAGE AND GOALS**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='possession team1', y='number of goals team1', data=a, label='Team1', color='blue')
sns.scatterplot(x='possession team2', y='number of goals team2', data=a, label='Team2', color='green')
plt.title('Possession Percentage vs Goals Scored')
plt.xlabel('Possession Percentage')
plt.ylabel('Goals Scored')
plt.legend()
plt.show()
```
